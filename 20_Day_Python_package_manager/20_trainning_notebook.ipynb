```python
def find_most_common_words(lines:list,num_of_words:int)-> any:
    most_common_words={}

    for line in lines:
        words = line.split(" ")
        for word in words:

            if word not in most_common_words:
                most_common_words[word] = 0
            else:
                most_common_words[word] += 1

    most_common_words_ordered = sorted(most_common_words.items(), key=lambda x:x[1], reverse=True)
    return most_common_words_ordered[0:num_of_words]
```

```python
find_most_common_words(text_list,30)
```
